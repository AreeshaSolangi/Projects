## handle Age column
- I will not delete nulls in the age column because it would delete 20% data which is huge
- since sknew is near 0 and mean, 50% percentile arent too far I will probably replace data with mean
- replacing 20% data with mean wont be bias

```python
df["Age"]=df.groupby("Gender")["Age"].transform(lambda x: x.fillna(x.mean()))
df["Age"]
```

```python
df["Age"]=df["Age"].astype(int)
```

# Univariate Analysis

```python
fig, axes= plt.subplots(3,3, figsize=(12,12))
axes=axes.ravel()
for i, predictor in enumerate(df.drop(columns=["Survived","Cabin","Fare","Ticket","Age","Name","PassengerId"])):
    sns.countplot(data=df, x=predictor, hue="Survived", ax=axes[i])
    plt.tight_layout()
    axes[i].legend(["Not Survived","Survived"])
    axes[i].set_title(f"Survival vs {predictor}")
axes[6].tick_params(axis='x', rotation=90);
```

### Insights From Univariate Analysis survival vs Individual Features
- Class: Upper class survived more, lower class less.
- Gender: Females survived more than males.
- Embarked: Cherbourg passengers survived more, Southampton less.
- Fare: Higher fare â†’ higher survival.
- Age: Children/teens survived more, adults/seniors less.
